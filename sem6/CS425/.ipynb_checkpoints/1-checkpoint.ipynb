```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
plt.plot(np.log10(np.array([0.01, 4, 8, 14, 18, 25, 39, 44, 71])), np.array([-30, -51, -58, -62, -65, -68, -74, -76, -84]))
plt.show()
```

```python
# sample points 
X = np.log10(np.array([0.01, 4, 8, 14, 18, 25, 39, 44, 71]))
Y = np.array([-30, -51, -58, -62, -65, -68, -74, -76, -84])

# solve for a and b
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.10f} + {:.10f}x'.format(a, b))

    return a, b

# solution
a, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
plt.plot(X, Y)
yfit = [a + b * xi for xi in X]
plt.plot(X, yfit)
```

```python
var=0
for i in (0,8):
    var = var + ((Y[i]-yfit[i])*(Y[i]-yfit[i]))
var/9
```
